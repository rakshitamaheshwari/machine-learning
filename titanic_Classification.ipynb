```python
#1 :avg=38..if age is null then check corresponding pclass and replace with avg value
#2: avg=29
#3: avg=25
```

```python
def change_cols(cols):                 #function replaces null values of age with their avg values acc to Pclass
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
        return Age
    
    
```

```python
dataset['Age']=dataset[['Age','Pclass']].apply(change_cols,axis=1)  #apply fun sends the data of pclass and age rowwise to function change_cols
#update age col acc to new values
```

```python
dataset.head(5)
```

```python
sns.heatmap(dataset.isnull())      #blackbox means no null values,now data is cleaned and ready to feed to ML model.
```

```python
dataset.dropna(inplace=True)  #agr thode bhot null rh gye h to drop them it will not affect our prediction
```

```python
gender=dataset['Sex']
Embarked=dataset['Embarked']
```

```python
Sex=pd.get_dummies(gender,drop_first=True)    #one-hot-encoding by pandas
Embarked=pd.get_dummies(Embarked,drop_first=True)
```

```python
dataset.head(5)  
```

```python
#cols who define a categorical variable,have to be made as dummies. SO as Pclass here.
pclass=dataset['Pclass']
Pclass=pd.get_dummies(pclass,drop_first=True)
```

```python
dataset=pd.concat([dataset,Sex,Embarked,Pclass],axis=1)
```

```python
dataset.head(1)
```

```python
dataset.drop(['PassengerId','Pclass','Name','Sex','Ticket','Fare','Embarked'],axis=1,inplace=True)
```

```python
y=dataset.iloc[:,0]
```

```python
X=dataset.iloc[:,1:]
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
